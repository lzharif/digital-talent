```python
# Data Source: https://data.go.id/dataset/data-upaya-kesehatan-per-provinsi

# Load the data first
import pandas as pd
import numpy as np

data = pd.read_csv('processed-data-upaya-kesehatan-per-provinsi-2000-2012.csv', delimiter=",", encoding="utf-8")
data.sample(10)
```

```python
# Check if she has a weakness (null values)
print("Amount of rows: ", data.nama_provinsi.count())
data.isnull().sum()
```

```python
# Nobody is perfect. She has 34 unwanted traits, especially at 'nilai' column.
# However, since her bad traits account for only 0.60% of all traits, we can forgive her and ignore it by dropping it
data = data.dropna()
print("Amount of rows after dropping the NaN: ", data.nilai.count())
```

```python
# Let's analyse her uniqueness
print("Unique value in each column:")
print(data.nunique())

# We find from her bio that there is 33 province in Indonesia, 13 health indicator in her. How about we list them?
print("Province in Indonesia:")
for province in data.nama_provinsi.unique():
    print("- ", province)
    
print("Health indicators:")
for indicator in data.indikator_upaya_kesehatan.unique():
    print("- ", indicator)
```

```python
# I'm curious about the health indicator. Is she single or not?
data.groupby('indikator_upaya_kesehatan')['nilai'].describe()
```

```python
# Well, it turns out that she has a unique personality ;)
# PersentasePeserta KB Baru looks quite strange though, as a percentage shouldn't value more than 100 (per cent)
# Now, let's compare each province by its health indicator
# In this notebook, I will compare the first five indicators only.

dataGroupingSum = data.groupby(['indikator_upaya_kesehatan', 'nama_provinsi'])['nilai'].sum()
dataGrouping = data.groupby(['indikator_upaya_kesehatan', 'nama_provinsi'])['nilai'].mean()

dataJPK = dataGroupingSum['Jumlah Peserta JPK']
dataBalitaGizi = dataGrouping['Persen Balita Gizi Buruk']
dataUCI = dataGrouping['Persen Desa UCI']
dataDropoutImunisasi = dataGrouping['Persen Drop out imunisasi']
dataImunisasiCampak = dataGrouping['Persen Imunisasi Campak']

# dataKBayi = dataGrouping['Persen Kunjungan Bayi']
# dataK1 = dataGrouping['Persen Kunjungan K1']
# dataK4 = dataGrouping['Persen Kunjungan K4']
# dataKN1 = dataGrouping['Persen Kunjungan Neonatus KN1']
# dataFe3 = dataGrouping['Persen Pemberian Fe3 Ibu Hamil']
# dataSalinakes = dataGrouping['Persen Salinakes']
# dataKbBaru = dataGrouping['PersentasePeserta KB Baru']
# dataBBLR = dataGrouping['Prevalensi BBLR per 100']
```

```python
# Let's compare 'Jumlah Peserta JPK'
# Looks like JPK user mainly comes from Prov. Jawa Tengah
graph1 = dataJPK.plot(kind='bar', title ="JPK User", legend=True, fontsize=12)
graph1.set_xlabel("Province", fontsize=12)
graph1.set_ylabel("Sum", fontsize=12)
```

```python
# How about 'Persen Balita Gizi Buruk'?
# It seems malnourished toddlers percentage is not quite big, with largest mean happen in Prov. Gorontalo for approximately 4%
graph2 = dataBalitaGizi.plot(kind='bar', title ="Malnourished Toddlers", legend=True, fontsize=12)
graph2.set_xlabel("Province", fontsize=12)
graph2.set_ylabel("Average (%)", fontsize=12)
```

```python
# Now, let's analyse 'Persen Desa UCI'!
# UCI or Universal Child Immunization seems to become norm although not perfect
# Three lowest UCI rate occured in Prov. Papua Barat, Sulawesi Barat, and Maluku Utara
graph3 = dataUCI.plot(kind='bar', title ="UCI Village Percentage", legend=True, fontsize=12)
graph3.set_xlabel("Province", fontsize=12)
graph3.set_ylabel("Average (%)", fontsize=12)
```

```python
# Now, let's analyze 'Persen Drop out imunisasi'!
# Three highest Immunization Dropout rate occured in Prov. NAD, Papua, and Sumatera Barat
graph4 = dataDropoutImunisasi.plot(kind='bar', title ="Immunization Dropout Rate", legend=True, fontsize=12)
graph4.set_xlabel("Province", fontsize=12)
graph4.set_ylabel("Average (%)", fontsize=12)
```

```python
# Now, let's analyze 'Persen Imunisasi Campak'!
# Compared to UCI, Measles immunisation rate is higher and resembles an early success
# However, Prov. Papua Barat and Sulawesi Barat still in lowest rate.
# One funny note: Jakarta has average percentage more than 100 %.
graph5 = dataImunisasiCampak.plot(kind='bar', title ="Measles Immunization Rate", legend=True, fontsize=12)
graph5.set_xlabel("Province", fontsize=12)
graph5.set_ylabel("Average (%)", fontsize=12)
```

## Conclusion
Based on the data, it seems that health indicator rate still low, especially in some remote province, e.g. Papua Barat, Maluku Utara, and Sulawesi Barat.
