```python
# Stratified Sampling function
def stratified_sampling(df, var, n_samples):
    data = df.groupby(var, group_keys=False).apply(lambda x: x.sample(min(len(x), n_samples)))
    return data
```

```python
print(df.groupby('Passholder Type')['Trip ID'].nunique())

# Clear null data
df = df.dropna()

# See distribution of passenger type
print(df.groupby('Passholder Type')['Trip ID'].nunique())
```

```python
# Because lowest variable comes from Staff Annual, we set the sampling by 300
# Sampling by passenger type
df_pass = stratified_sampling(df, 'Passholder Type', 300)
df_pass.head()
```

```python
# Count distance by kilometres
lon1, lat1, lon2, lat2 = map(np.radians, [df['Starting Station Longitude'], df['Starting Station Latitude'],df['Ending Station Longitude'], df['Ending Station Latitude']])
dlat = lat2 - lat1 
dlon = lon2 - lon1
a = (np.sin(dlat/2))**2 +np.cos(lat1) * np.cos(lat2) * (np.sin(dlon/2))**2 
c = 2 * np.arcsin( np.sqrt(a)) 
d = 6373  * c 
df_pass['distance in km']=d
df_pass['distance in km'].hist(bins=100,range=(0,5))

# Describe by group
```
