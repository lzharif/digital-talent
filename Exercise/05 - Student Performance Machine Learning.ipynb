```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from collections import defaultdict

import scipy.stats as stats
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier, plot_importance
from sklearn.model_selection import train_test_split

df = pd.read_csv("xAPI-Edu-Data.csv", sep=",")
print(df.shape)
df.head(20)
```

```python
# Check null values
col_names = df.columns
num_data = df.shape[0]
for c in col_names:
    num_non = df[c].isin(["?"]).sum()
    if num_non > 0:
        print (c)
        print (num_non)
        print ("{0:.2f}%".format(float(num_non) / num_data * 100))
```

```python
# labeling all categorical variable
d = defaultdict(LabelEncoder)

# Encoding the attribute variable
X_var = df.drop("Class", axis = 1)
y = df["Class"]
X = X_var.apply(lambda x: d[x.name].fit_transform(x) if x.dtype == "O" else x)

# # Inverse the encoded
# fit.apply(lambda x: d[x.name].inverse_transform(x) if x.dtype == "O" else x)

# # Using the dictionary to label future data
# df.apply(lambda x: d[x.name].transform(x) if x.dtype == "O" else x)
X.head(20)
```

## Undersampling Data

```python
y_m,y_h,y_l = y.value_counts()
df_m = X.copy()
df_m['Class'] = y
df_m.head()
df_class_l = df_m[df_m['Class'] == 'L']
df_class_m = df_m[df_m['Class'] == 'M']
df_class_h = df_m[df_m['Class'] == 'H']
```

```python
df_class_m_low = df_class_m.sample(y_l)
df_class_h_low = df_class_h.sample(y_l)
df_bal = pd.concat([df_class_l, df_class_m_low, df_class_h_low], axis = 0)
print(df_bal['Class'].value_counts())

X_bal = df_bal.drop("Class", axis = 1)
y_bal = df_bal["Class"]
X_train, X_test, y_train, y_test = train_test_split(X_bal, y_bal, test_size= 0.2)
```

```python
keys = []
scores = []
devia = []
models = {'Logistic Regression': LogisticRegression(), 'Decision Tree': DecisionTreeClassifier(),
          'Random Forest': RandomForestClassifier(n_estimators=300, random_state=52),
         'Linear SVM': SVC(kernel = 'linear'), 'RBF SVM': SVC(kernel = 'rbf'),
         'Naive Bayes': MultinomialNB(), 'XGBoost': XGBClassifier()}

for k,v in models.items():
    mod = v
    cross_val = cross_val_score(mod, X_bal, y_bal, cv=10)
#     mod.fit(X_train, y_train)
#     pred = mod.predict(X_test)
#     print('Results for: ' + str(k) + '\n')
#     print(confusion_matrix(y_test, pred))
#     print(classification_report(y_test, pred))
#     acc = accuracy_score(y_test, pred)
#     print("accuracy is "+ str(acc)) 
#     print('\n' + '\n')
    keys.append(k)
#     scores.append(acc)
    scores.append(cross_val.mean())
    devia.append(cross_val.std())
    table = pd.DataFrame({'model':keys, 'accuracy score':scores, 'standard deviation':devia})

print(table)
```

```python
# Exploring the number of estimators in the random forest
import seaborn as sns
score = []
est = []
estimators = [1, 10, 50, 100, 200, 300, 400, 500]
for e in estimators:
    rfc1 = RandomForestClassifier(n_estimators=e, random_state=52)
    pred1 = rfc1.fit(X_train, y_train).predict(X_test)
    accuracy = accuracy_score(y_test, pred1)
    score.append(accuracy)
    est.append(e)
plot = sns.pointplot(x=est, y=score)
plot.set(xlabel='Jumlah estimator', ylabel='Akurasi', 
         title='Akurasi berdasarkan jumlah estimator')
plt.show()
```

```python
# Exploring minimum leaf samples
score = []
leaf = []
leaf_options = [1, 5, 10, 50, 100, 200]
for l in leaf_options:
    rfc2 = RandomForestClassifier(n_estimators=400, random_state=52, min_samples_leaf=l)
    pred2 = rfc2.fit(X_train, y_train).predict(X_test)
    accuracy = accuracy_score(y_test, pred2)
    score.append(accuracy)
    leaf.append(l)
plot = sns.pointplot(x=leaf, y=score)
plot.set(xlabel='Jumlah minimum leaf samples', ylabel='Akurasi', 
         title='Akurasi berdasarkan jumlah minimum leaf score')
plt.show()
```

```python
from sklearn import metrics
rfc = RandomForestClassifier(n_estimators=400, min_samples_leaf=1, random_state=52)
rfc.fit(X_train,y_train)
y_train_predicted = rfc.predict(X_train)
y_test_predicted = rfc.predict(X_test)
train_accuracy = metrics.accuracy_score(y_train, y_train_predicted)
test_accuracy = metrics.accuracy_score(y_test, y_test_predicted)

print("Train Accuracy = " + str(train_accuracy))
print("Test Accuracy = " + str(test_accuracy))
```

```python
from sklearn.preprocessing import OneHotEncoder
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam

# One Hot encode the class labels
y_onehot = pd.get_dummies(y)
# encoder = OneHotEncoder(sparse=False)
# y_onehot = encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size= 0.2)

# define baseline model
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(64, activation='softmax'))
    model.add(Dense(32, activation='softmax'))
    model.add(Dropout(0.1))
    model.add(Dense(16, activation='softmax'))
    model.add(Dense(y_onehot.shape[1], activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

model = baseline_model()
checkpointer = ModelCheckpoint(filepath="weights.hdf5", verbose=1, save_best_only=True, monitor='loss', mode='min')
model.fit(X_train, y_train, verbose=2, batch_size=5, epochs=200, callbacks=[checkpointer])

model.load_weights('weights.hdf5')
```

```python
results = model.evaluate(X_test, y_test)

print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```
