```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# Join ada 4 Jenis:

- Inner Join
- Left Join
- Right Join
- Cross Join

Terdapat Join ke diri Sendiri yang disebut Self Join

## 1. INNER JOIN

<img src="http://www.mysqltutorial.org/wp-content/uploads/2011/05/MySQL-INNER-JOIN-Venn-Diagram-300x166.png" />

<img src="http://www.mysqltutorial.org/wp-content/uploads/2017/07/MySQL-Join-inner-join-illustration.png" />

<img src="http://www.mysqltutorial.org/wp-content/uploads/2011/05/products_and_product_lines_tables.png" />

```python
run("""
SELECT 
    productCode, 
    productName, 
    textDescription
FROM
    products t1
        INNER JOIN
    productlines t2 ON t1.productline = t2.productline
""")
```

## USING

```python
run("""
SELECT 
    productCode, 
    productName, 
    textDescription
FROM
    products
        INNER JOIN
    productlines USING (productline)
""")
```

## Kombinasi dengan GROUP BY

<img src="http://www.mysqltutorial.org/wp-content/uploads/2011/05/orders_order_details_tables.png" />

```python
run("""
SELECT 
    T1.orderNumber,
    status,
    SUM(quantityOrdered * priceEach) total
FROM
    orders T1
        INNER JOIN
    orderdetails T2 ON T1.orderNumber = T2.orderNumber
GROUP BY orderNumber
""") # bisa pakai AS atau gak
```

## Join dengan Operasi Perbandingan Lain

```python
run("""
SELECT 
    orderNumber, 
    productName, 
    msrp, 
    priceEach
FROM
    products p
        INNER JOIN
    orderdetails o ON p.productcode = o.productcode
        AND p.msrp > o.priceEach
WHERE
    p.productcode = 'S10_1678'
""")
```

```python
run("""
SELECT
 c.customerNumber,
 c.customerName,
 orderNumber,
 o.status
FROM
 customers c
LEFT JOIN orders o ON c.customerNumber = o.customerNumber
""")
```

```python
run("""
SELECT 
    c.customerNumber, 
    c.customerName, 
    orderNumber, 
    o.status
FROM
    customers c
        LEFT JOIN
    orders o ON c.customerNumber = o.customerNumber
WHERE
    orderNumber IS NULL
""")
```

## WHERE vs ON

```python
run("""
SELECT 
    o.orderNumber, 
    customerNumber, 
    productCode
FROM
    orders o
        LEFT JOIN
    orderDetails USING (orderNumber)
WHERE
    orderNumber = 10123
""") # WHERE dilakukan terakhir
```

```python
run("""
SELECT 
    o.orderNumber, 
    customerNumber, 
    productCode
FROM
    orders o
        LEFT JOIN
    orderDetails d ON o.orderNumber = d.orderNumber
        AND o.orderNumber = 10123
""") # Karena pada left join, semua harus ditampilkan tanpa pandang bulu. Yang tidak sesuai ON, ditampilkan NULL
```

# 3. RIGHT JOIN

<img src="http://www.mysqltutorial.org/wp-content/uploads/2017/07/MySQL-Join-right-join-illustration.png" />

```python
run("""
SELECT 
    concat(e.firstName,' ', e.lastName) salesman, 
    e.jobTitle, 
    customerName
FROM
    employees e
        RIGHT JOIN
    customers c ON e.employeeNumber = c.salesRepEmployeeNumber
        AND e.jobTitle = 'Sales Rep'
ORDER BY customerName
""")
```

```python
run("""
SELECT 
    concat(e.firstName,' ', e.lastName) salesman, 
    e.jobTitle, 
    customerName
FROM
    employees e
        RIGHT JOIN
    customers c ON e.employeeNumber = c.salesRepEmployeeNumber
        AND e.jobTitle = 'Sales Rep'
ORDER BY customerName
""")
```

# 5. Self Join

```python
run("""
SELECT 
    CONCAT(m.lastname, ', ', m.firstname) AS 'Manager',
    CONCAT(e.lastname, ', ', e.firstname) AS 'Direct report'
FROM
    employees e
        INNER JOIN
    employees m ON m.employeeNumber = e.reportsto
ORDER BY manager
""")
```

```python
run("""
SELECT 
    IFNULL(CONCAT(m.lastname, ', ', m.firstname),
            'Top Manager') AS 'Manager',
    CONCAT(e.lastname, ', ', e.firstname) AS 'Direct report'
FROM
    employees e
        LEFT JOIN
    employees m ON m.employeeNumber = e.reportsto
ORDER BY manager DESC
""")
```

```python
run("""
SELECT 
    c1.city, c1.customerName, c2.customerName
FROM
    customers c1
        INNER JOIN
    customers c2 ON c1.city = c2.city
        AND c1.customername > c2.customerName
ORDER BY c1.city
""")
```

```python
# One to many
# 1. Report the account representative for each customer.
run("""
SELECT
 c.customerNumber,
 c.customerName,
 CONCAT(e.firstname, ' ', e.lastname) AS accountRepresentative,
 e.jobTitle
FROM
 customers c
LEFT JOIN employees e ON c.salesRepEmployeeNumber = e.employeeNumber
""")
```

```python
# One to many
# 2. Report total payments for Atelier graphique.

run("""
SELECT 
    SUM(p.amount)
FROM
    customers c
        INNER JOIN
    payments p USING(customerNumber)
WHERE
    c.customerName = 'Atelier graphique'
""")
# run("SELECT * FROM customers")
```

```python
# One to many
# 3. Report the total payments by date

run("SELECT paymentDate,SUM(amount) AS TotalPayment FROM payments GROUP BY paymentDate")
```

```python
# One to many
# 4. Report the products that have not been sold.
run("SELECT * FROM orderDetails")
run("""
SELECT 
    p.productName,
    o.quantityOrdered
FROM
    products p
        LEFT JOIN
    orderdetails o USING(productcode)
WHERE o.quantityOrdered IS NULL
""")
```

```python
# 1. List products sold by order date.
run("""
SELECT orderDate,productName,quantityOrdered
FROM orders
    JOIN orderDetails USING (orderNumber)
    JOIN products USING (productCode)
ORDER BY orderDate""")
```

```python
# One to many
# 8. What is the quantity on hand for products listed on 'On Hold' orders?
run("""
SELECT productName, quantityOrdered, quantityInStock
FROM orders
    JOIN orderdetails USING(orderNumber)
    JOIN products USING(productCode)
WHERE status = 'On Hold'
""")
```
